#### The Data Science Handbook Chapter Summary
*Chapter 1 - DJ Patil*

DJ Patil worked in research and consulting prior to joining the tech world. He co-coined the term "Data Scientist" and built one of the first premier data science teams. He was appointed as the first Chief Data Scientist of the United States.

He talks about taking risks when starting in the field of data science. There is no clear-cut definition of what data science is, or what a data scientist does. So, a good way to start is to prove that you can do or build *something*. Next comes the task of communicating or describing what you built to other people. I think this is where pedagogy comes in.

Another key point is that data science is a team sport. As good as someone may be, success in the field usually comes as a result of collaboration between people with different backgrounds and skillsets. Like a whole being greater than the sum of its parts.

Next, DJ gives his opinion on how to develop one's skills as a data scientist. He says we do not know for sure what works or what doesn't, but it is very important that you put in your hours. Learn new skills, work with other people, be around individuals who are the best in the world.

The last advise I like best: Work on simple things. Simple things become hard; hard things become intractable. 

*Chapter 2 - Hilary Mason*

Hilary Mason is the founder of Fast Forward Labs, a machine intelligence learning company and is the Data Scientist in Residence at Accel. Previously, she was the Chief Scientist at bitly.

She describes her day-to-day work at Accel: She helps companies in their portfolio navigate interesting or challenging data questions.

She says data science is a rising trend. Few people are doing it, but it is clearly applicable to most domains and not just in tech.

The next five years will see a lot of change. Understanding will be greater, things will be more defined, tooling (should) be better, and data science will be an integral part of daily work.

Hilary also talks about the importance of communication and storytelling. I see it as bridging the gap between tech and non-tech people to get a better understanding of the problem and the approach in solving it.

She also thinks that startups are a good place to develop skills in data science. In this environment, it is easier to get your hands on good data sets and choose your own approach in analysing it. Otherwise, it is important that you work with and for people who you are comfortable with, and who understand what they want you to do.

She also shared some guide questions when working on a data science project:
1. What are you working on?
2. How will you know when you're done?
3. What does it impact?

Having answers to these questions will ensure you're making good use of your time.

Next, she talks about the importance of learning and playing on your strengths. For fresh graduates, smaller companies are better especially if you have a good mentor who can teach you the right skills. 

Data science work can be divided into three buckets: Stats, Code, Visualisation. For aspiring data scientists, it is a good idea to work on projects that highlight your strengths. Then, another project where you're worst at. This will help you grow and round up your skills.

Lastly, she encourages people to put up their work on GitHub, and just be persistent as they learn and develop their skills in data science.
