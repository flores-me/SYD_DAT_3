*Linear Regression*

Wikipedia defines [Linear Regresssion (LR)](www.https://en.wikipedia.org/wiki/Linear_regression) as an "approach for modeling the relationship between a scalar dependent variable *y* and one or more explanatory variables *X*."

Simply put, it is a means of predicting the value of something (let's call this something *y*) given the state of one or more related things (let's collectively call these related thing(s) *X*).

Take a person's height at age 18, for instance (they say we stop growing at 18). Intuitively, a person's "final height" is influenced by a number of factors: Generally, males are taller than females. Genetics also plays a major role-- if both parents are tall or short then we tend to be the same. There can be more but for purposes of discussion I'll stop here.

In this example, *y* is a person's height at age 18, while *X* are: (1) Gender (2) Height of father and (3) Height of mother.

In order to do this prediction, LR needs a set of actual *y* values and their associated *X* values as input in order to build the prediction model. If I were to be part of this set, my entry would look like:

*Height:* 171cm; *Gender:* Male; *Height of father:* 162cm; *Height of mother:* 150cm

The set should contain a significant number of entries so an accurate model can be created. For our example, let's say that there are 99 more entries. 

For each record in the set, LR looks at the other 3 attributes and makes an assumption on how these 3 produce the height. This assumption is the prediction model. Think of it as a formula or rule of thumb for predicting the value of *y*. The 3 chosen attributes may or may not help in making a good prediction. LR has a way of "rating" how each of the attributes actually help making a prediction. This "rating" is called the *p-value*. The lower, the better.

Let's pretend that given the set, LR came up with the assumption: "If female, take the average of parents' heights and subtract 5cm. If male, take the average of parents' heights and add 15cm.". Given this logic, the model can now make a prediction on a person's height given the gender, mother's height, and father's height of someone who was not part of the initial survey.

Let this someone be my sister:

*Height:* 152cm; *Gender:* Female; *Height of father:* 162cm; *Height of mother:* 150cm

Given the model, her predicted height will be: ((162+150)/2) - 5 = 151cm. This is a pretty good prediction. The percentage of "good" predictions a certain model will make is called *R-squared* in statistics. This value is a number between 0 and 1. The higher, the better.

Basing from my sister's sample above, the chosen attributes and the model are fairly accurate. The *p-value* for all 3 must be small and the *R-squared* value must be high.

* Choosing Features *

There are 3 classical methods for selecting predictors in a model. To summarise:

1. *Forward Selection* -  You start with a null model (i.e. no predictors). You then add predictors one at a time until you satisfy a "stopping rule".

2. *Backward Selection* - You start with *all* predictors in the model. Observe the p-value of all variables and remove the variable with the highest p-value (i.e. least statistically). Repeat until you satisfy a "stopping rule".

3. *Mixed Selection* - This is similar to forward selection. You start with a null model and add predictors incrementally. However, at each step you evaluate the p-values and may opt to remove a predictor that falls outside a defined threshold. Repeat until you satisfy a "stopping rule".
